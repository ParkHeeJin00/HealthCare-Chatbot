```python
import pandas as pd
df = pd.read_csv('./data/rawal_data.csv')
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import nltk
nltk.download('punkt')

model_dir = "lcw99/t5-base-korean-text-summary"
tokenizer = AutoTokenizer.from_pretrained(model_dir)
model = AutoModelForSeq2SeqLM.from_pretrained(model_dir)

max_input_length = 100000
q = df['question'][50]
inputs_q = ["summarize: " + q]
a = df['answer'][50]
inputs_a = ["summarize: " + a]

inputs = tokenizer(inputs_a, max_length=max_input_length, truncation=True, return_tensors="pt")
output = model.generate(**inputs, num_beams=8, do_sample=True, min_length=10, max_length=400)
decoded_output = tokenizer.batch_decode(output, skip_special_tokens=True)[0]
predicted_title = nltk.sent_tokenize(decoded_output.strip())[0]
print("원문 : ", a)
print("요약문 : ", predicted_title)

inputs = tokenizer(inputs_q, max_length=max_input_length, truncation=True, return_tensors="pt")
output = model.generate(**inputs, num_beams=8, do_sample=True, min_length=10, max_length=100)
decoded_output = tokenizer.batch_decode(output, skip_special_tokens=True)[0]
predicted_title = nltk.sent_tokenize(decoded_output.strip())[0]
print("원문 : ", q)
print("요약문 : ", predicted_title)
```

```python
df = pd.read_csv('./data/rawal_data.csv')
```

```python
max_input_length = 100000

a = max(df['question'])
inputs_a = ["summarize: " + a]

inputs = tokenizer(inputs_a, max_length=max_input_length, truncation=True, return_tensors="pt")
output = model.generate(**inputs, num_beams=8, do_sample=True, min_length=10, max_length=360)
decoded_output = tokenizer.batch_decode(output, skip_special_tokens=True)[0]
predicted_title = nltk.sent_tokenize(decoded_output.strip())[0]
print("원문 : ", a)
print("요약문 : ", predicted_title)
```

```python
# question, answer 을 입력으로 받아서, question, answer를 요약한 df를 생성

def summarize_text(text):
    inputs = ["summarize: " + text]
    inputs = tokenizer(inputs, max_length=max_input_length, truncation=True, return_tensors="pt")
    output = model.generate(**inputs, num_beams=8, do_sample=True, min_length=10, max_length=100)
    decoded_output = tokenizer.batch_decode(output, skip_special_tokens=True)[0]
    predicted_title = nltk.sent_tokenize(decoded_output.strip())[0]
    return predicted_title

df['question'] = df['question'].apply(summarize_text)
df['answer'] = df['answer'].apply(summarize_text)

df.to_csv('./data/summarized_data.csv', index=False)
```

```python
# df question 중에 있는 문자열 중에 가장 긴 행의 인덱스 구하기
max_len = 0
max_index = 0
for i in range(len(df['question'])):
    if len(df['question'][i]) > max_len:
        max_len = len(df['question'][i])
        max_index = i

print(max_index)
```
